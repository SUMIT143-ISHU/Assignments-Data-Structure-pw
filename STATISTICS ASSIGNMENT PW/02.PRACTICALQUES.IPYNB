Q2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1  # Lower bound of the distribution
high = 6  # Upper bound of the distribution

# Generate random samples from the discrete uniform distribution
num_samples = 1000  # Number of random samples
random_samples = np.random.randint(low, high+1, size=num_samples)

# Calculate the PMF
values, counts = np.unique(random_samples, return_counts=True)
probabilities = counts / num_samples

# Plot the PMF
plt.figure(figsize=(8, 6))
plt.bar(values, probabilities, color='skyblue', edgecolor='black')
plt.title('Probability Mass Function (PMF) of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(values)
plt.ylim(0, max(probabilities) + 0.1)
plt.show()
```

Q3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
def bernoulli_pdf(x, p):
    """
    Calculate the Probability Density Function (PDF) for a Bernoulli distribution.
    
    Parameters:
    x (int): The outcome, should be either 0 or 1.
    p (float): The probability of success (0 <= p <= 1).
    
    Returns:
    float: The probability of the outcome x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1")
    if p < 0 or p > 1:
        raise ValueError("p must be a value between 0 and 1")
    
    if x == 1:
        return p
    elif x == 0:
        return 1 - p

# Example usage:
p = 0.7  # Probability of success
x = 1     # Outcome (could also be 0)

# Calculate the PDF for x = 1
pdf_value = bernoulli_pdf(x, p)
print(f"The PDF for x = {x} with p = {p} is: {pdf_value}")

# Calculate the PDF for x = 0
pdf_value = bernoulli_pdf(0, p)
print(f"The PDF for x = 0 with p = {p} is: {pdf_value}")
```

Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1  # Lower bound of the distribution
high = 6  # Upper bound of the distribution

# Generate possible outcomes for the discrete uniform distribution
outcomes = np.arange(low, high + 1)

# The probability of each outcome in a discrete uniform distribution
probability = 1 / len(outcomes)  # Each outcome has the same probability

# Calculate the CDF by accumulating the probabilities
cdf_values = np.cumsum(np.repeat(probability, len(outcomes)))

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.step(outcomes, cdf_values, where='post', label='CDF', color='blue', linewidth=2)

# Customize the plot
plt.title(f'Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.xticks(outcomes)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(True)
plt.show()
```

Q9.Write a Python function to calculate Z-scores from a dataset and plot them?

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate the Z-scores for a dataset.

    Parameters:
    data (array-like): The dataset (list or numpy array).

    Returns:
    numpy array: The Z-scores for the dataset.
    """
    # Calculate the mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-scores using the formula
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(0, 1, 1000)  # Generate 1000 random samples from a normal distribution

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(8, 6))
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

# Add a title and labels
plt.title('Z-scores of Dataset')
plt.xlabel('Z-score')
plt.ylabel('Density')

# Display the grid and show the plot
plt.grid(True)
plt.show()
```

Q10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 10000  # Number of population data points
sample_size = 50  # Size of each sample
num_samples = 1000  # Number of samples to take

# Generate data from a non-normal distribution (Uniform distribution)
population_data = np.random.uniform(0, 10, population_size)

# Function to simulate the Central Limit Theorem
def simulate_clt(population_data, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        # Take a random sample from the population
        sample = np.random.choice(population_data, size=sample_size, replace=True)
        # Compute the mean of the sample
        sample_means.append(np.mean(sample))
    return sample_means

# Simulate the CLT
sample_means = simulate_clt(population_data, sample_size, num_samples)

# Plot the population distribution and the sampling distribution (sample means)
plt.figure(figsize=(12, 6))

# Plot the population distribution
plt.subplot(1, 2, 1)
plt.hist(population_data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Population Distribution (Uniform)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, alpha=0.7, color='salmon', edgecolor='black', density=True)
plt.title('Sampling Distribution of the Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

Q17.Generate random variables and calculate their corresponding probabilities using the binomial distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def generate_binomial_distribution(n, p, size):
    """
    Generate random variables from a binomial distribution and calculate their corresponding probabilities.
    
    Parameters:
    - n: Number of trials
    - p: Probability of success in each trial
    - size: Number of random variables to generate
    
    Returns:
    - random_variables: The generated random variables (successes in trials)
    - probabilities: Corresponding probabilities of the generated random variables
    """
    # Step 1: Generate random variables from the binomial distribution
    random_variables = np.random.binomial(n, p, size)
    
    # Step 2: Calculate the probability for each random variable
    probabilities = binom.pmf(random_variables, n, p)
    
    return random_variables, probabilities

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success
size = 1000  # Number of random variables to generate

# Generate random variables and their corresponding probabilities
random_variables, probabilities = generate_binomial_distribution(n, p, size)

# Plot histogram of random variables (observed values)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(random_variables, bins=np.arange(n+2)-0.5, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Variables from Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')

# Plot the probabilities corresponding to the random variables
plt.subplot(1, 2, 2)
plt.scatter(random_variables, probabilities, alpha=0.7, color='salmon')
plt.title('Probability of Each Random Variable')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')

# Show the plots
plt.tight_layout()
plt.show()
```

Q18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(X, mu, sigma):
    """
    Calculate the Z-score for a given data point.
    
    Parameters:
    - X: Data point
    - mu: Mean of the distribution
    - sigma: Standard deviation of the distribution
    
    Returns:
    - Z-score
    """
    Z = (X - mu) / sigma
    return Z

def plot_standard_normal_distribution():
    """
    Plot the standard normal distribution for comparison.
    """
    # Create a range of x values for the standard normal distribution (mean=0, std=1)
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)  # Standard normal distribution

    # Plot the standard normal distribution
    plt.plot(x, y, label="Standard Normal Distribution", color='b', lw=2)
    plt.title('Standard Normal Distribution (Mean=0, Std=1)')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Parameters for the normal distribution
mu = 50  # Mean of the distribution
sigma = 10  # Standard deviation of the distribution

# Given data point
X = 60  # Example data point

# Calculate the Z-score for the given data point
z_score = calculate_z_score(X, mu, sigma)

# Compare the Z-score with the standard normal distribution
probability = norm.cdf(z_score)  # CDF of the standard normal distribution

# Print the results
print(f"Z-score for X = {X} is: {z_score:.2f}")
print(f"The cumulative probability for Z = {z_score:.2f} is: {probability:.4f}")

# Plot the standard normal distribution
plot_standard_normal_distribution()

# Mark the Z-score on the plot
plt.axvline(z_score, color='r', linestyle='--', label=f'Z = {z_score:.2f}')
plt.legend()
plt.show()
```

Q.20 Create a confidence interval for a dataset using Python and interpret the result?

```python
import numpy as np
import scipy.stats as st

# Sample data (replace with your actual data)
data = [25, 28, 22, 27, 24, 26, 23, 29, 25, 28]

# Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation

# Confidence level (e.g., 95%)
confidence_level = 0.95

# Calculate degrees of freedom
degrees_of_freedom = len(data) - 1

# Calculate t-critical value
t_critical = st.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

# Calculate margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(len(data)))

# Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")

# Interpretation
print(f"We are {confidence_level * 100}% confident that the true population mean falls within the interval [{lower_bound:.2f}, {upper_bound:.2f}].")
```
